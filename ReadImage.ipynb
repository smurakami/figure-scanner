```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

from figure import Figure
```

# 図形を読み込んで頂点に分割

```python
figure = Figure(cv2.imread('okamura.png'))
plt.imshow(figure.image)
plt.scatter(figure.vertex[:, 1], figure.vertex[:, 0])
```

# 各頂点の外角を取得

これが、どれだけ曲がっているかという動きの特徴量になる。

```python
plt.plot(figure.angle)
```

ちなみに一周しているので、外角を積分するとだいたい`2*pi`になる。(今回は右回りなのでマイナス)

```python
plt.plot(np.cumsum(figure.angle))
```
